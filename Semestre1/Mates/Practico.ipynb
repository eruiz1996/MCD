# Ecuación del calor

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Parámetros

```python
L = 10 #Longitud del dominio
T = 2 #Tiempo final
k = 0.1 #Coeficiente de difusión térmica
```

### Condiciones iniciales y de **frontera**.

Aquí algo profe.

Yo he visto en la literatura que se usa más el término frontera en lugar de contorno.

Esto por lo que significa en áreas como el Análisis Matemático y la Topología, al referirnos a la frontera nos referimos aquellos puntos que no importa la que tan pequeña sea el radio, la bola que se hace (términos topológicos) toma elementos en el abierto.

```python
# inicial
def u0(x):
    return np.sin(np.pi*x/L)
# frontera
def uL(t):
    return 0
```

Esto si no mal dicen mis notas de EDP es un problema **inicial con valores en la frontera homogéneos** y a la condición de frontera se le dice **condición de Dirichlet**.

### Método de las diferencias finitas

**Discretización**

Justo y necesario para hacer las gráficas en `matplotlib` además de ser el paso uno del método de las diferencias finitas.

Yo resolví esta ecuación pero con Fourier, entonces éste método numérico fue nuevo para mí.

```python
Nx = 100 #Número de puntos en el espacio
Nt = 500 #Número de saltos temporales

dx = L / Nx
dt = T / Nt

# aquí es donde se hace la discretización para la visualización
x = np.linspace(0, L, Nx)
t = np.linspace(0, T, Nt)

# matriz de ceros con la forma (puntos-espacio, puntos-tiempo)
u = np.zeros((Nx, Nt))
# llenamos la matriz
u[:,0] = u0(x)
```

Notar que con `u0(x)` se está aplicando, de manera vectorizada, la condición inicial al vector `x`

```python
u0(x)
```

**Aplicación del método**

Sobre esto encontré que se está aplicando de manera explícita, pues se resuelve al tiempo $t+1$ usando $u(x,t)$

```python
r = k * dt / dx ** 2

for n in range(Nt-1):
    u[0, n+1] = uL(t[n+1])
    for i in range(1, Nx-1):
        u[i, n+1] = u[i, n] + r*(u[i+1,n] - 2*u[i,n] + u[i-1,n])
```

**Visualización**

```python
X,T=np.meshgrid(x,t)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
surf=ax.plot_surface(X,T,u.T,cmap='viridis')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Posición')
ax.set_zlabel('Temperatura')
ax.set_title('Ecuación de calor')
plt.show()
```

## Interpretación
La ecuación del calor aquí planteada nos habla de la conducción térmica del calor en un intervalo finito, lo que es en términos más coloquiales es que nos modela la propagación del calor en un material.

En la gráfica podemos apreciar lo siguiente:

1. Las condicones de frontera hacen que en los extremos valga 0 $u(0,t)=u(L,t)=0$ (condición de Diritchlet) y esto sería equivalente a decir que el material está totalmente aislado (térmicamente hablando) en sus extremos.

2. La condición inical senoidal nos da muestra la temperatura al tiempo ( $t = 0$ ), esto nos dice que hay una distribución de calor con una forma de onda (de ahí la forma que tiene la gráfica) en el centro del material, que luego se va disipando con el tiempo debido a la difusión térmica.

# Ecuación de onda

### Parámetros

```python
L = 10 #Longitud del dominio
T = 2 #Tiempo límite
c = 10 #Velocidad de onda
```

### Condiciones iniciales y de frontera
Son las mismas, entonces no las volveré a declarar.

Sé que se me puede decir que las variables `L` y `T` también  son iguales, sin embargo ahora tienen otra interpretación.

### Método de las diferencias finitas
**Discretización**

```python
Nx = 100 #Puntos en el espacio
Nt = 500 #Pasos en el tiempo
dx = L / Nx
dt = T / Nt

x = np.linspace(0, L, Nx)
t=np.linspace(0, T, Nt)

u = np.zeros((Nx, Nt))
u[:,0] = u0(x)
```

**Aplicación del método**

Se hace explícito.

Aquí interesante, en realidad tanto el método anterior como este son explícitos porque ambos usan el $u(x,t)$ para modelar el siguiente instante. Investigué y dicen que es fácil distinguir entre un modelo implícito y uno explícito, ya que en el implícito se necesita resolver un sistema de ecuaciones.

```python
r = c * dt/dx

for n in range(Nt-1):
    u[0, n+1] = uL(t[n+1])
    for i in range(1, Nx-1):
        u[i,n+1] = 2*(1-r**2)*u[i,n] - u[i,n-1] + r**2*(u[i+1,n]+u[i-1,n])
```

**Visualización**

```python
X, T = np.meshgrid(x,t)
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
surf = ax.plot_surface(X, T, u.T, cmap='viridis')
ax.set_xlabel('Posición')
ax.set_ylabel('Tiempo')
ax.set_zlabel('Desplazamiento')
ax.set_title('Ecuación de onda')
plt.show()
```

## Interpretación
Aquí de las condiciones de Dirichlet se puede suponer que una cuerda está atada en sus dos extremos y por eso en ambos vale cero.

Lo de la condición inicial se puede interpretar de una manera más curiosa, ya que estaríamos diciendo que la cuerda incialmente tiene una forma senoidal. Esto si nos ponemos a pensar tiene mucho sentido porque cuando se tira una cuerda nunca cae derecha, siempre parece como una víbora.

Finalmente, la ecuación de onda modela la propagación de una onda en un material (en nuestra interpretación, una cuerda). Sobre este tema creo yo que el más interesante es el de la resonancia (la vibración que todos los objetos tenemos de manera natural), ya que se tienen casos como el punte de Tacoma.

Dato curioso: a raíz del suceso en el puente de Tacoma se les prohibió a los soldados marchar cuando pasen por puentes, ya que esto puede ocasionar que entren en resonancia, haya amplitud de onda y todo quiebre.

# Ecuación de Laplace

### Dominio

```python
L = 10
Nx = 100
Ny = 200

x = np.linspace(0, L, Nx)
y = np.linspace(0, L, Ny)
X, Y = np.meshgrid(x, y)
```

### Condiciones iniciales y de frontera

```python
u = np.zeros((Nx,Ny))

u[:, 0] = 100 #Condición de contorno
u[:, -1] = 0
u[0, :] = 50
u[-1, :] = 75
```

### Método de relajación

```python
iteraciones = 1_000 # dato curioso, en Python puedes separar los número con _ como si fueran separador de miles
tolerancia = 1e-6 # bien profe!

for iteracion in range(iteraciones):
    u_anterior = np.copy(u)
    for i in range(1, Nx-1):
        for j in range(1, Ny-1):
            u[i,j] = 0.25*(u_anterior[i+1,j] + u_anterior[i-1,j] + u_anterior[i,j+1] + u_anterior[i,j-1])

    #Verificamos la convergencia
    if np.max(np.abs(u-u_anterior)) < tolerancia:
        break
```

**Visualización**

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,u.T)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Valor de u')
ax.set_title('Ecuación de Laplace')
plt.show()
```

### Descripción/interpretación
De esta ecuación no sabía nada... investigué y dice que es una ecuación que modela estados en equilibrio.

Por lo que entiendo en las condiciones de frontera, se somenten a diferentes temperaturas en cada esquina, entonces deduzco que la gráfica nos estaría mostrando el proceso que se tomó para llegar al equilibrio.

Es como si el calor viajara hacia el interior y se fuera curvando con base en las temperaturas de los contornos.

Lo que sí que estudié fue la Transforma de Laplace y sobre eso cuando quiera hablamos que para mí usa uno de las heurísticas matemáticas más importantes (lo digo con hechos): un problema complicado de resolver se pasa a un espacio donde es más fácil de manejar (¿a quién no le recuerda eso una biyección?), se resuelve el problema en ese espacio y luego se regresa al espacio original.

# Ecuación de Poisson

### Parámetros

```python
L = 10
Nx = 100
Ny = 100
```

### Definición de la malla

```python
x = np.linspace(0, L, Nx)
y = np.linspace(0, L, Ny)

X, Y = np.meshgrid(x, y)
```

### Definición de la función fuente

```python
f = np.sin(np.pi*X/L)**3*np.cos(np.pi*Y/L)
```

### Condiciones de frontera

```python
#Condiciones de contorno (de Dirichlet): lo sabía!!
u = np.zeros((Nx, Ny))
u[:,0] = 0
u[:,-1] = 0
u[0,:] = 0
u[-1,:] = 0
```

### Método de las diferencias finitas

```python
max_iterations = 1000
tolerance = 1e-6

for iteration in range(max_iterations):
    u_old = np.copy(u)
    for i in range(1,Nx-1):
        for j in range(1,Ny-1):
            u[i,j] = 0.25*(u_old[i+1,j] + u_old[i-1,j] + u_old[i,j+1] + u_old[i,j-1] - f[i,j])

    #Verificación de convergencia:
    if np.max(np.abs(u-u_old))<tolerance:
        break
```

**Visualización**

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,u.T)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Valor de u')
ax.set_title('Ecuación de Poisson')
plt.show()
```

### Descripción/interpretación
Por lo que vi esta ecuación es generalizar la función de Laplace ya que si la función fuente es igual a cero, entonces se obtiene esta ecuación. Ahora, debido a que se tienen condiciones de Dirichlet podemos decir que toda la frontera de la función está fija en 0 y entonces  va a ser diferente interpretación que la de Laplace, puesto que no hay temperaturas que vayan hacia el interior. De esto concluyo que todas las deformaciones (curvas) que vemos en la gráfica deben de ser atribuidas a la función fuente.

# Ecución de transporte

### Parámetros

```python
L = 10 #Longitud del dominio
T = 2 #Tiempo final
b = 1 #Velocidad
c = 0.1 #Coeficiente de difusión
```

### Condición incial

```python
def u0(x):
    return np.exp(-x**3)
```

### Método de diferencias finitas
**Discretización**

```python
Nx = 100
Nt = 100
dx = L/Nx
dt = T/Nt

x = np.linspace(0, L, Nx)
t = np.linspace(0, T, Nt)

u = np.zeros((Nx,Nt))
u[:,0] = u0(x)
```

**Aplicación del método**

```python
r = c * dt/dx**2

for n in range(Nt-1):
    for i in range(1,Nx-1):
        u[i,n+1] = u[i,n] - b*dt/(2*dx)*(u[i,n]-u[i-1,n]) + r*(u[i+1,n] - 2*u[i,n] + u[i-1,n])
```

**Visualización**

```python
X, T = np.meshgrid(x,t)
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X,T,u.T)
ax.set_xlabel('Posición')
ax.set_ylabel('Tiempo')
ax.set_zlabel('Valor de u')
ax.set_title('Ecuación de transporte')
plt.show()
```

### Descripción/interpretación
Esta ecuación modela dos cosas:
- El desplazamiento de una sustancia (por eso se dice del transporte)
- Tendencia del suavizamiento por difusión.
Al ser una condición inicial una función exponecial, podemos decir que da valores no negativos (se ve gráficamente) y tampoco simétricos.

Es interesante, esto me recuerda a las cortaduras de nivel para describir gráficas tridimensionales por medio de cortes. Si hacemos un corte en el momento $t=0$ podemos ver claramente la función exponencial (identificada claramente por la cola pesada). A medida que pasa el tiempo se aprecia cómo la ecuación va bajando ($t=1, 2, \ldots$) y supongo yo que a esto se refiere la tendencia al suavizamiento.

# Ecuación de Burgers

### Parámetros

```python
L = 10
T = 5
nu = 10
Nx = 100
Nt = 500
```

### Condiciones inciales

```python
u = np.zeros((Nx,Nt))
u[:, 0] = np.exp(-np.pi*x/L)
```

### Método de las diferencias finitas centradas
**Discretización**

```python
dx = L/Nx
dt = T/Nt
x = np.linspace(0, L, Nx)
t = np.linspace(0, T, Nt)
```

### Aplicación del método

```python
r = nu*dt/(2*dx)

for n in range(Nt-1):
    for i in range(1,Nx-1):
        u[i,n+1] = u[i,n] - u[i,n]*(dt/(2*dx))*(u[i+1,n] - u[i-1,n]) + r*(u[i+1,n] - 2*u[i,n] + u[i-1,n])
```

**Visualización**

```python
X, T = np.meshgrid(x,t)
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
surf = ax.plot_surface(X,T,u.T,cmap='viridis')
ax.set_xlabel('Espacio')
ax.set_ylabel('Tiempo')
ax.set_zlabel('Valor de u')
ax.set_title('Ecuación de Burgers resuelta con diferencias finitas')
plt.show()
```

### Descripción/interpretación
Esta ecuación modela 3 cosas!
- Transporte
- Difusión
- No linealidad
De condiciones iniciales tenemos nuevamente una función exponencial...
